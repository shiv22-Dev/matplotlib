```python
#plt.plot(x,y,color = 'color name',linestyle='line_style',Linewidth=value,marker='marker symbol',lebel='Label name')
#plt.xlabel('text')
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1000,1500,1200,1000]

plt.plot(months, sales,
         color='blue',
         linestyle='--',
         linewidth=2,
         marker='o',
         label='2025 sales data')

plt.xlabel('sales per Months')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months, sales, color='blue', linestyle='--', linewidth=2, marker='o', label='2025 sales data')

plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Data Report')

plt.legend(loc='upper left', fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=1)

plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])

plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt

regions = ['North','south','east','west']
revenue = [3000,2000,1500,1000]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','coral'])
plt.title("revenue contribution by region")
plt.show()
```

```python
##plt.subplot(nrows,ncols,index)

import matplotlib.pyplot as plt
x= [1,2,3,4]
y= [10,20,30,40]

plt.subplot(1,2,1)#1 row ,2nd column ,2 subplot
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2)#1row,2colomn,2nd subplot
plt.bar(x,y)
plt.title('Bar chart')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2, figsize=(10,5))

x = [1,2,3,4]
y = [10,15,20,25]
ax[1].bar(x, y, color='blue')
ax[0].set_title('Bar Chart')

ax[1].bar(x, y, color='green')
ax[1].set_title('Bar Chart')

fig.suptitle('Comprision of line and bar charts')
plt.tight_layout()
plt.show()
```
